```python
import pandas as pd
import requests
import io

def baixar_e_abrir_csv(url, sep=';', encoding='latin1'):

        #Pega url
        response = requests.get(url)

        # Decodifica o conteúdo com a codificação correta
        arquivo_buffer = io.StringIO(response.content.decode(encoding))

        # Lê o conteúdo como CSV
        df = pd.read_csv(arquivo_buffer, sep=sep)

        return df
```

```python
'''
1. Variáveis principais

Estimativa Café (Conab): previsões de safra (produção total, área plantada, produtividade).
Série Histórica Café (Conab): produção passada, volumes e rendimentos.
Preços minimos café (Conab-Preços Mínimos)
Preços mensal e semanais café por estado

Dólar (BCB): taxa de câmbio influencia diretamente exportações.
IPCA (IBGE): inflação geral da economia.
Custos de importação/exportação (SECEX, MDIC): logística, tarifas e frete internacional.

-- Preços dentro do processo produtivo



2. Outras variáveis

Preços internacionais do café: ICE (Nova York para arábica, Londres para robusta).
Consumo interno e externo: dados da OIC (Organização Internacional do Café).
Clima: chuvas, temperatura, geadas, seca (INMET, CPTEC, NOAA). FEITO
Estoque de passagem: volume armazenado ao final de cada ano-safra.
Custos de produção agrícola (Conab/Embrapa): insumos, mão de obra, defensivos.
Logística interna: preço do diesel, custo do transporte até os portos.
'''
```
